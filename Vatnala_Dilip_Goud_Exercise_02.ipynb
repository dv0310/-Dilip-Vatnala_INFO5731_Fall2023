```python
import requests
from bs4 import BeautifulSoup
import time

def scrape_google_scholar(query, num_pages):
    base_url = "https://scholar.google.com"
    headers = {
        "User-Agent": "Your User Agent Here",
        "accept-language": "en-US,en;q=0.9",
    }

    articles = []

    for page in range(0, num_pages):
        # Construct the search URL
        url = f"{base_url}/scholar?start={page * 10}&q={query}&hl=en"

        # Send a GET request to the search page
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "html.parser")

            # Extract article information
            for result in soup.select(".gs_ri"):
                title = result.select_one(".gs_rt a")
                venue = result.select_one(".gs_a")
                year = result.select_one(".gs_a")
                authors = result.select_one(".gs_a")
                abstract = result.select_one(".gs_rs")

                if title and venue and year and authors and abstract:
                    title = title.text
                    venue = venue.text.split('-')[1].strip()
                    year = year.text.split('-')[0].strip()
                    authors = authors.text.split('-')[0].strip()
                    abstract = abstract.text

                    articles.append({
                        "Title": title,
                        "Venue": venue,
                        "Year": year,
                        "Authors": authors,
                        "Abstract": abstract
                    })

            time.sleep(2)  # Be polite and avoid overloading the server

    return articles

# Search for articles on Google Scholar
query = "information retrieval"
num_pages = 10  # Adjust the number of pages as needed

articles = scrape_google_scholar(query, num_pages)

# Print the collected articles
for i, article in enumerate(articles, start=1):
    print(f"Article {i}:")
    print(f"Title: {article['Title']}")
    print(f"Venue: {article['Venue']}")
    print(f"Year: {article['Year']}")
    print(f"Authors: {article['Authors']}")
    print(f"Abstract: {article['Abstract']}")
    print("\n")
```
