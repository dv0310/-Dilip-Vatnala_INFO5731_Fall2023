Question 2 (10 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction.

```python
# Import necessary libraries
from six.moves import urllib  # Using the six library for compatibility
import pandas as pd  # Import pandas for data manipulation

# Initialize an empty list to store text data
text_data = []

# Define the URL of the text file to be downloaded
file_url = "https://raw.githubusercontent.com/dv0310/new-sample-txt/main/sampledata.txt"

# Open the URL to retrieve data
response = urllib.request.urlopen(file_url)

# Iterate over each line in the response
for line in response:
    # Decode the binary data to UTF-8 and remove newline characters
    decoded_line = line.decode("utf-8").replace('\r\n', '')

    # Check if the line is not empty
    if decoded_line:
        # Append the decoded line to the text_data list
        text_data.append(decoded_line)

# Create a DataFrame from the collected text data using pandas
df = pd.DataFrame(text_data, columns=['Text'])

# Print the DataFrame
print(df)
```

```python
#To determine the number of sentences in a data frame

df['Num of sentences'] = df['Text'].apply(lambda y: len(str(y).split(".")))
df
```

```python
#To determine the number of words in a data set

df['Num of Words'] = df['Text'].apply(lambda y: len(str(y).split(" ")))
df
```

```python
#To determine the number of characters in a data frame

df['Num of Char'] = df['Text'].str.len()
df
```

```python
#To determine the list's average word length


def awl(x):
    jai = x.split() #here the sentence is being splitted into parts and being assigned to variable jai
    if len(jai) != 0:  #here we are checking if the variable is empty or not
        return(sum(len(deep) for deep in jai)/len(jai))
    else:
        return None

df['Avg Word Length'] = df['Text'].apply(lambda z: awl(z))
df
```

```python
#To determine the number of special characters in the list

def spchar(x):
    red = 0
    for i in range(len(x)):
        if not((x[i].isalpha()) and (x[i].isdigit())):
            red = red + 1
    return red

df['Num of spec char'] = df['Text'].apply(lambda y: spchar(y))
df
```

```python
#To determine the amount of numerics in the list

df['Num of num'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.isdigit()])) #here we are considering it to be a seperate numeric if it is seperated with a space using the split function
df
```

```python
#To determine the amount of uppercase words in a list

df['Num of upper case words'] = df['Text'].apply(lambda x: len([x for x in x.split() if x.isupper()])) #using split function we are finding the words with completely upper case
df
```

```python
df.to_csv('data.csv',index=True)
```

```python
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity

# URL to sample text data
sample_text_url = "https://raw.githubusercontent.com/dv0310/new-sample-txt/main/sampledata.txt"

# Query to search for in the text data
search_query = "This is a query document."

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")

# Tokenize the search query and text data
query_tokens = tokenizer.tokenize(search_query)
text_tokens_list = [tokenizer.tokenize(text) for text in text_data]

# Convert tokens to their corresponding token IDs
query_ids = tokenizer.convert_tokens_to_ids(query_tokens)
text_ids_list = [tokenizer.convert_tokens_to_ids(tokens) for tokens in text_tokens_list]

# Add [CLS] and [SEP] tokens and convert to tensors
query_ids = torch.tensor([tokenizer.build_inputs_with_special_tokens(query_ids)])
text_ids_list = [torch.tensor([tokenizer.build_inputs_with_special_tokens(ids)]) for ids in text_ids_list]

# Generate embeddings for the search query and text data
with torch.no_grad():
    query_embedding = model(query_ids)[0].mean(dim=1)  # Calculate mean-pooled embeddings for the query
    text_embeddings_list = [model(ids)[0].mean(dim=1) for ids in text_ids_list]

# Calculate cosine similarity between the query and each text document
similarities = [cosine_similarity(query_embedding, text_embedding)[0][0] for text_embedding in text_embeddings_list]

# Rank the text documents based on similarity in descending order
ranked_documents = sorted(enumerate(text_data), key=lambda x: similarities[x[0]], reverse=True)

# Print the ranked documents and their similarities
for rank, (idx, text) in enumerate(ranked_documents, 1):
    print(f"Rank {rank}: Similarity = {similarities[idx]:.4f}, Text = {text}")
```
